```python
import requests
r = requests.get('https://www.dcard.tw/_api/forums/job/posts?popular=true')
response = r.text
response
```

```python
import json
js = json.loads(response)
js
```

```python
title = []
for i in js:
    title.append(i['title'])
title
```

```python
content = []
for j in js:
    content.append(j['excerpt'])
content
```

```python
time = []
for t in js:
    time.append(t['createdAt'])
time
```

```python
img = []
img_jpg = []
img_png = []
img_else = []
for m in js:
    for k in m['mediaMeta']:
        img.append(k['url'])
img
```

```python
import pandas as pd
df = pd.DataFrame({'Title': title , 'Content': content , 'Time': time})
df
```

```python
# 下載檔案
import urllib.request
import zipfile
```

```python

for i in range(len(img)):
    name="image"+str(i)+".jpg"
    with open(name, 'wb') as handler:
        img_data = urllib.request.urlretrieve(img[i],filename = name)
        handler.write(img_data)
```
